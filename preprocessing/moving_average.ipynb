```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/weather_underground.csv").set_index("date")
df.shape
```

```python
plt.plot(df.index, df.maxdewpti)
```

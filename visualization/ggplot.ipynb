```python
from ggplot import *
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 100)
y = x * 4 + np.random.normal(0, 3, 100)

df = pd.DataFrame({"x" : x, "y" : y})
```

```python
g = ggplot(df, aes("x", "y")) \
     + geom_point(color='red') \
     + geom_line(color='red') \
     + ggtitle("title") \
     + xlab("haha") \
     + ylab("good")
print(g)
```

```python
from ggplot import *
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

##### Scatter and Line Plot

```python
x = np.linspace(0, 10, 100)
y = x * 3 + np.random.normal(0, 5, 100)
z = np.array(range(100)) % 2
df = pd.DataFrame({"x" : x, "y" : y, "z" : z})
```

```python
g = ggplot(df, aes("x", "y")) \
     + geom_point(color='red') \
     + geom_line(color='red') \
     + ggtitle("title") \
     + xlab("haha") \
     + ylab("good")
print(g)
```

##### Density Plot
* Plot maximum likelihood estimate of a p=0.7 biased coin.
* Use this plot to turn histogram into density.

```python
samples = np.random.binomial(n=100, p=0.7, size=10000)
p_mles = samples / 100
```

```python
df = pd.DataFrame({"trials" : p_mles})
df.dropna().shape
```

```python
g = (ggplot(df, aes("trials")) \
     + geom_density(alpha = 0.5) \
     + xlab("Prior") \
     + ylab("Count"))
print(g)
```

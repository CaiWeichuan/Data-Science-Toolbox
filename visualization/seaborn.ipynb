```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# create fake data
mu = [3, 5, 4, 8]
sigma = [0.1, 0.3, 0.2, 0.3]
data = [np.random.normal(mu[i], sigma[i], 1000) for i in range(len(mu))]

# combine column 2, 4 to create column 5
data.append(data[1] / 2 + data[3])
labels = ["x%i"%(i+1) for i in range(len(data))]

df = pd.DataFrame(data, index = labels)
df = df.T
```

### Use distplot to plot single column

```python
sns.distplot(df.x1)
```

### Use violinplot to examine spread

```python
sns.violinplot(data=df)
```

### Use pairplot to examine pair-wise correlation

```python
sns.pairplot(df, diag_kind='kde', size=2.4);
```

### Use heatmap for a more quantitative view of the correlation matrix

```python
sns.heatmap(df.corr())
```

### Use clustermap to dissect large correlation matrix

```python
sns.clustermap(df.corr())
```

### Examine strongly correlated feature pair

```python
sns.jointplot(df.x4, df.x5, df, kind='reg')
```

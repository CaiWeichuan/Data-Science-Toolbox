```python
import numpy as np
from matplotlib import pyplot as plt
```

#### Line plot

```python
plt.figure(figsize=(8, 4))

# create sample data
x = np.linspace(- np.pi, 2 * np.pi, 100)
sinx = np.sin(x)
cosx = np.cos(x)

# plot two lines
plt.plot(x, sinx)
plt.plot(x, cosx)

# plot vertical， horizontal lines
plt.axhline(0, linestyle = "--", color = 'grey')
plt.axvline(0, linestyle = ":", color = 'grey')

plt.plot([0], [0], 'o', markersize=10, color='b')
plt.plot([np.pi], [np.cos(np.pi)], 'v', markersize=10, color='g')
plt.plot([np.pi/2], [np.sin(np.pi/2)], '^', markersize=10, color='r')

# annotation
plt.legend(["sine(x)", "cos(x)"], frameon=False)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Sample Plot")

# turn off splines
for spine in plt.gca().spines.values():
    spine.set_visible(False)
    
# get rid of axis
plt.axis('off')
    
plt.show()
```

#### Scatter 

```python
plt.figure(figsize=(8, 4))

# create sample data
x1 = np.random.normal(10, 0.2, 1000)
y1 = np.random.normal(5, 1, 1000)

x2 = np.random.normal(8, 0.8, 1000)
y2 = np.random.normal(6, 0.3, 1000)

# set axis limit
plt.xlim([5, 12])
plt.ylim([0, 10])

# alternative: set axis as [x_low, x_high, y_low, y_high]
# plt.axis([5, 12, 0, 10])

plt.grid(True)

plt.scatter(x1, y1, alpha=0.3, marker="v")
plt.scatter(x2, y2, alpha=0.3, marker="o")

# turn off splines
for spine in plt.gca().spines.values():
    spine.set_visible(False)
    
# get rid of axis
plt.axis('off')

plt.show()
```

#### Histogram

```python
plt.figure(figsize=(8, 4))

# create sample data
sigma = 1
mu = 5
x = np.random.normal(mu, sigma, 1000)

# divide into bins
count, bins, ignored = plt.hist(x, 30, density=True, facecolor = 'g', alpha = 0.4)

# plot histogram and probability density
plt.plot(bins, 
         1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2)), 
         linewidth=2, color='r')

plt.grid(True)

# use math expression in title
plt.title('$\sigma_s=%i$, $\mu_s=%i$'%(sigma, mu))

# turn off splines
for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.show()
```

```python
plt.figure(figsize=(8, 4))

# create sample data
x1 = np.random.normal(1, 1, 1000)
x2 = np.random.normal(5, 3, 1000)

# divide into bins
count, bins, ignored = plt.hist([x1, x2], 30, density = True, alpha = 0.4, label = ["x1", "x2"])

plt.grid(True)
plt.legend(loc='upper left', frameon=False)

# turn off splines
for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.show()
```

#### Pie chart

```python
plt.figure(figsize=(8, 8))

# values must be numbers, labels must be strings
values = np.linspace(0, 80, 5)
labels = [str(val) for val in values]

_ = plt.pie(x = values, 
            labels = labels,
            autopct='%1.1f%%',     # add percentage
            labeldistance = 1.1)   # percentage tag distance from center

# draw a circle at center
centre_circle = plt.Circle((0,0),0.70,fc='white')
plt.gca().add_artist(centre_circle)

plt.show()
```

#### Bar Chart

```python
# Enter raw data
aluminum = np.array([6.4e-5 , 3.01e-5 , 2.36e-5, 3.0e-5, 7.0e-5, 4.5e-5, 3.8e-5, 4.2e-5, 2.62e-5, 3.6e-5])
copper = np.array([4.5e-5 , 1.97e-5 , 1.6e-5, 1.97e-5, 4.0e-5, 2.4e-5, 1.9e-5, 2.41e-5 , 1.85e-5, 3.3e-5 ])
steel = np.array([3.3e-5 , 1.2e-5 , 0.9e-5, 1.2e-5, 1.3e-5, 1.6e-5, 1.4e-5, 1.58e-5, 1.32e-5 , 2.1e-5])
```

```python
# Calculate the average
aluminum_mean = np.mean(aluminum)
copper_mean = np.mean(copper)
steel_mean = np.mean(steel)

# Calculate the standard deviation
aluminum_std = np.std(aluminum)
copper_std = np.std(copper)
steel_std = np.std(steel)
```

```python
# Create lists for the plot
materials = ['Aluminum', 'Copper', 'Steel']
x_pos = np.arange(len(materials))
CTEs = [aluminum_mean, copper_mean, steel_mean]
error = [aluminum_std, copper_std, steel_std]
```

```python
# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, CTEs, yerr=error, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Coefficient of Thermal Expansion ($\degree C^{-1}$)')
ax.set_xticks(x_pos)
ax.set_xticklabels(materials)
ax.set_title('Coefficent of Thermal Expansion (CTE) of Three Metals')
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.show()
```

#### Arrow 

```python
fig, ax = plt.subplots()

x_pos = [0, -0.5]
y_pos = [0, 0.5]
x_direct = [1, 0]
y_direct = [1, -1]

ax.quiver(x_pos, y_pos, x_direct, y_direct)
ax.axis([-1.5, 1.5, -1.5, 1.5])

plt.show()
```

#### Subplot

```python
plt.figure(figsize=(8, 8))

# subplot setup
nplots = 4
ncols = 2
nrows = 2
colors = ['b', 'r', 'g', 'm']

# create fake data
mu = [0, 1, 4, 8]
sigma = [0.1, 3, 0.02, 10]
data = [np.random.normal(mu[i], sigma[i], 1000) for i in range(nplots)]

for i in range(nplots):
    x = data[i]

    # parameter: number of rows, number of columns, plot index
    ax = plt.subplot(nrows, ncols, i + 1)
    
    # can be any type of plot (histogram, pie, line etc.)
    ax.hist(x, 50, density=True, color=colors[i], alpha = 0.4)
    
    ax.set_ylabel("Density")
    ax.set_title("Plot %i"%i)

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["bottom"].set_visible(False)

ax.invert_yaxis()

# use tight layout so texts don't overlap
plt.tight_layout()
```

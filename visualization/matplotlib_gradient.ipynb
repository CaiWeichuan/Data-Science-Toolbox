```python
import numpy as np
from matplotlib import pyplot as plt
```

#### Meshgrid

The quiver arrow color argument needs to have the same dimensions as the position and direction arrays

```python
x = np.arange(0, 2 * np.pi + 2 * np.pi / 20, 2 * np.pi / 20)
y = np.arange(0, 2 * np.pi + 2 * np.pi / 20, 2 * np.pi / 20)

X, Y = np.meshgrid(x, y)

u = np.sin(X) * np.cos(Y)
v = - np.cos(X) * np.sin(Y)

fig, ax = plt.subplots()

ax.quiver(X, Y, u, v)

ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.axis([0, 2 * np.pi, 0, 2 * np.pi])
ax.set_aspect('equal')

plt.show()
```

```python
x = np.arange(0, 2.2, 0.2)
y = np.arange(0, 2.2, 0.2)

X, Y = np.meshgrid(x, y)
u = np.cos(X) * Y
v = np.sin(y) * Y

n = - 2
color_array = np.sqrt(((v - n) / 2) ** 2 + ((u - n) / 2) ** 2)
print(color_array.shape)

fig, ax = plt.subplots()
ax.quiver(X, Y, u, v, color_array, alpha=0.8)

ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.axis([-0.2, 2.3, -0.2, 2.3])
ax.set_aspect('equal')

plt.show()
```

#### Gradient

```python
x = np.arange(-2, 2.2, 0.2)
y = np.arange(-2, 2.2, 0.2)

X, Y = np.meshgrid(x, y)
z = X * np.exp(- X ** 2 - Y ** 2)

dx, dy = np.gradient(z)

n = - 990
color_array = np.sqrt(((dx - n) / 2) ** 2 + ((dy - n)/2)**2)
print(color_array.shape)

fig, ax = plt.subplots()
ax.quiver(X, Y, dx, dy, color_array)

ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_aspect('equal')

plt.show()
```

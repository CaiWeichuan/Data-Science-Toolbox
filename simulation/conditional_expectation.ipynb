```python
import matplotlib.pyplot as plt
```

#### Non-geometric
Roll out experiment, What’s the expected waiting time to receive treatment? 1K users
* Day 1: 10 users out of 1000 -> trt
* Day 2: 10 users out of 990 -> trt
* Day 3: 10 users out of 980 -> trt
* …
* Day 100: 10 users -> trt


```python
pds = [10/1000]
pis = [10/1000]
pd_ = 1 - pds[-1]
ewt = [0]
for i in range(2, 101):
    pi = 10 / (1000 - (i - 1) * 10)
    pd = pd_ * pi
    pd_ = pd_ * (1 - pi)
    
    pds.append(pd)
    pis.append(pi)
    ewt.append(pd * i)
```

```python
plt.figure(figsize=(10, 4))

ax = plt.subplot(1, 2, 1)
ax.plot(pds)
ax.set_title("unconditional probability of being chosen vs day")

ax = plt.subplot(1, 2, 2)
ax.plot(pis)
ax.plot(ewt)
ax.set_title("expected weight time vs day")
ax.legend(["conditional prob", "expected weight time"])

plt.tight_layout()
plt.show()
```

```python
# sanity check, probability should sum to 1
sum(pds)
```

```python
# expected weight time
sum(ewt)
```

#### Geometric
Roll out experiment: What’s the expected waiting time for the user to first enroll in the treatment? 1K users
* Day 1: 10 users out of 1000 -> trt
* Day 2: 10 users out of 1000 -> trt
* Day 3: 10 users out of 1000 -> trt
* …
* Day 1000: 10 users -> trt

```python
ewt = [] # expected weight time on day 1, 2, ... 1000
pds = [] # probability of being chosen on day 1, 2... 1000 conditioned on haven't been chosen

p = 10 / 1000
p_ = 1 - p

pis = [p] * 1000

for i in range(1, 1001):
    pd = p_ ** (i - 1) * p 
    pds.append(pd)
    ewt.append(pd * i)
```

```python
# sanity check, probability should sum to 1
sum(pds)
```

```python
# expected weight time
sum(ewt)
```

```python
plt.figure(figsize=(10, 4))

ax = plt.subplot(1, 2, 1)
ax.plot(pds)
ax.set_title("unconditional probability of being chosen vs day")

ax = plt.subplot(1, 2, 2)
ax.plot(pis)
ax.plot(ewt)
ax.set_title("expected weight time vs day")
ax.legend(["conditional prob", "expected weight time"])

plt.show()
```

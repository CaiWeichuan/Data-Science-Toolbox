```python
import numpy as np
import matplotlib.pyplot as plt
```

#### PDF using inversion method

```python
u = np.random.uniform(low=0, high=1, size=1000000)
x = np.log(u / (1 - u))
```

```python
plt.hist(x, bins=1000, density=True, alpha=0.5)
plt.hist(np.random.normal(0, 1, 1000000), bins=1000, density=True, alpha=0.5)
plt.grid()
plt.xlim(-10, 10)
plt.legend(['Logistic PDF', "Unit normal PDF"])
plt.show()
```

```python
np.mean(x)
```

```python
np.std(x)
```

#### CDF empirical vs. analytical

```python
x_arr = np.linspace(-10, 10, 10000)
```

```python
y = 1 / (1 + np.exp(- x_arr))
```

```python
plt.hist(x, bins=1000, cumulative=True, density=True, alpha=0.5)
plt.plot(x_arr, y, color='r')
plt.xlim(-10, 10)
plt.axvline(0, alpha=0.5, color='b')
plt.axhline(0.5, alpha=0.5, color='b')
plt.legend(["CDF analytical", "CDF empirical"])
plt.grid()
plt.show()
```

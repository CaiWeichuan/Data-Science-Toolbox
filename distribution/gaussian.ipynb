```python
from scipy.stats import norm
import numpy as np
from matplotlib import pyplot as plt
```

#### Draw Sample

```python
sample = np.random.normal(loc=14, scale=3, size=1000)
plt.hist(sample, label='norm pdf')
```

```python
sample = norm.rvs(loc=4, scale=3, size=1000, random_state=None)
plt.hist(sample, label='norm pdf')
```

#### Draw PDF

```python
x = np.linspace(-10, 10, 10000)
rv = norm()
plt.plot(x, rv.pdf(x), label='norm pdf')
```

```python
plt.plot(x, norm.pdf(x, loc=3, scale=1))
```

```python
plt.plot(x, norm.cdf(x, loc=3, scale=1))
```

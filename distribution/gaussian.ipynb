```python
from scipy.stats import norm
import numpy as np
from matplotlib import pyplot as plt
```

#### Draw Sample

```python
sample = np.random.normal(loc=14, scale=3, size=1000)
plt.hist(sample, label='norm pdf')
```

#### Draw PDF

```python
x = np.linspace(-10, 10, 10000)
```

```python
# create a random variable
rv = norm()
plt.plot(x, rv.pdf(x), label='norm pdf')
```

```python
# apply analytical formula
plt.plot(x, norm.pdf(x, loc=3, scale=1))
```

```python
# apply analytical formula
plt.plot(x, norm.cdf(x, loc=3, scale=1))
```

#### Plot a region

```python
# define constants
mu = 998.8 
sigma = 73.10
x1 = 900
x2 = 1100
```

```python
# calculate the z-transform
z1 = ( x1 - mu ) / sigma
z2 = ( x2 - mu ) / sigma
```

```python
# build the plot
fig, ax = plt.subplots()

x = np.arange(z1, z2, 0.001) # range of x in spec
y = norm.pdf(x,0,1)
ax.fill_between(x, y, 0, alpha=0.3, color='b')

x_all = np.arange(-10, 10, 0.001)
y2 = norm.pdf(x_all,0,1)
ax.fill_between(x_all, y2, 0, alpha=0.1)


ax.plot(x_all, y2)

ax.set_xlim([-4,4])
ax.set_yticklabels([])

plt.show()
```

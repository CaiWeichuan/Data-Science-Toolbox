```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial as fact
import scipy
```

```python
def plot_all(c, width, degrees, func, expand):
    plt.figure(figsize=(12, 8))

    x = np.linspace(c - width, c + width, 10000)
    y_exact = func(x)
    
    for degree in degrees:
        plt.plot(x, expand(x, degree, c))

    plt.plot(x, y_exact)
    plt.legend(["%i degree"%i for i in degrees] + ["Exact"])
    plt.show()
```

#### Taylor series
* Every smooth function can be approximated by a polynomial.

#### Expamd exponential function

```python
def exp_expand(x, n = 1, c = 0):
    y_est = 1
    for i in range(n):
        y_est += np.exp(c) * (x - c) ** (i + 1) / fact(i + 1)
    return y_est
```

Maclaurin series (c = 0)

```python
plot_all(c=0, width=5, degrees=[3, 5, 10, 12], func=np.exp, expand=exp_expand)
```

Expand at c = 11

```python
plot_all(c=11, width=5, degrees=[3, 5, 10, 12], func=np.exp, expand=exp_expand)
```

#### Expamd sine function

```python
def sin_expand(x, n = 1, c = 0):
    y_est = scipy.zeros_like(x)
    # cycle: sin, cos, -sin, -cos
    for i in range(n + 1):
        if i % 2 == 1:
            # odd-degree term
            y_est += (-1) ** (i // 2) * np.cos(c) * (x - c) ** (i) / fact(i)
        else:
            # even-degree term
            y_est += (-1) ** (i // 2) * np.sin(c) * (x - c) ** (i) / fact(i)
    return y_est
```

```python
plot_all(c=0, width=4, degrees=[3, 5, 10, 12], func=np.sin, expand=sin_expand)
```

```python
plot_all(c=np.pi, width=4, degrees=[3, 5, 10, 12], func=np.sin, expand=sin_expand)
```

#### Expamd cosine function

```python
def cos_expand(x, n=1, c=0):
    y_est = scipy.zeros_like(x)
    # cycle: cos, -sin, -cos, sin
    for i in range(n + 1):
        if i % 2 == 1:
            # odd-term
            y_est += (-1) ** ((i + 1) // 2) * np.sin(c) * (x - c) ** (i) / fact(i)
        else:
            # even-term
            y_est += (-1) ** ((i + 1) // 2) * np.cos(c) * (x - c) ** (i) / fact(i)
    return y_est
```

```python
plot_all(c=0, width=4, degrees=[3, 5, 10, 12], func=np.cos, expand=cos_expand)
```

```python
plot_all(c=4, width=4, degrees=[3, 5, 10, 12], func=np.cos, expand=cos_expand)
```
